```python
import themes

themes.register()
```

## The Data

```python
from themes import datasets

help(datasets.load_markets)
```

```python
markets = (
    datasets.load_markets()
    .pipe(lambda df: df[df["timestamp"] > "2020"])
    .assign(
        relative_price=lambda df: df.groupby("symbol")["close"].transform(
            lambda g: g / g.iloc[0] - 1
        )
    )
)

markets
```

## Plot with Altair

```python
import altair as alt
```

```python
with alt.themes.enable("capon"):
    chart = (
        alt.Chart(markets)
        .mark_line()
        .encode(
            x=alt.X("timestamp", title=None, axis=alt.Axis(format="%b %y")),
            y=alt.Y(
                "relative_price", title="Relative Price", axis=alt.Axis(format="+%")
            ),
            color="symbol",
            tooltip=[
                "timestamp",
                "symbol",
                alt.Tooltip("relative_price", format="+.2%"),
            ],
        )
        .properties(
            title={
                "text": f"Market Indexes Change",
                "subtitle": f"Relative to {markets['timestamp'].dt.date.min()}",
            },
            width=800,
            height=300,
        )
    )

    display(chart)
```

## Plot with Matplotlib

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'

plt.figure();   # Hack to overcome mpl bug of delaying with figure.facecolor update
```

```python
with plt.style.context("capon"):
    plt.figure(figsize=(12, 4), dpi=96)
    #     print(plt.rcParams['figure.facecolor'])
    markets.pivot_table(
        index="timestamp", columns="symbol", values="relative_price"
    ).plot(ax=plt.gca())
```
